```python
# replace ph levels with median of same age, icutype and gender

mask = ['age','gender','icutype']
fill_col = ['first_ph','last_ph','mean_ph','min_ph','max_ph']

new_rows = []

for ind, row in mortality.iterrows():
    
    age = row['age']
    gender = row['gender']
    icutype = row['icutype']
    
    median_row = filled_median_group.loc[(filled_median_group.index.get_level_values('age')==age)&\
                                    (filled_median_group.index.get_level_values('icutype')==icutype)&\
                                (filled_median_group.index.get_level_values('gender')==gender)]
    
    new_row = []
    
    for col in fill_col:
        
        new_val = median_row[col].get_values()
        
        for val in new_val:
        
            if row[col] > 14:

                 new_row.append(val)

            else:
                new_row.append(row[col])
    
    new_rows.append(new_row)

mortality_filled_out = pd.DataFrame(data=new_rows,columns = mortality[fill_col].columns,index=mortality[fill_col].index)
mortality_filled_out.head()
```

```python
class Categorical_Extractor(BaseEstimator, TransformerMixin):
    """Get dummy variables for categorical features"""
    
    def __init__(self):
        pass

    def get_dummy_(self,df):
        
        icu_dummy = pd.get_dummies(df['icutype'],drop_first=True)
        df = pd.concat((df['gender'],icu_dummy),axis=1)
        
        return df.values.reshape(-1, 4)
    
    def transform(self, X, *args):
        
        X = self.get_dummy_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
class Numerical_Extractor(BaseEstimator, TransformerMixin):
    """Use RobustScaler on numerical features"""
    
    def __init__(self):
        pass

    def robust_scale_(self,df,threshold=10):
        
        num_col = []
        for col, num in df.nunique().iteritems():
            if num > threshold:
                num_col.append(col)
                
        df = df[num_col]
        
        rscaler = RobustScaler()
        array = rscaler.fit_transform(df)
        
        return array.reshape(-1,array.shape[1])
    
    def transform(self, X, *args):
        
        X = self.robust_scale_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
class Clean_FeatureUnion(BaseEstimator, TransformerMixin):
    """Convert array from feature union into dataframe"""
    
    def __init__(self):
        pass
    
    
    def array_to_df(self,array):
        
        new_columns = list(X.columns)
        new_columns.extend(['icutype3','icutype4'])
        
        X_df = pd.DataFrame(array,columns=new_columns,index=X.index)
        X_df = X_df.rename(columns={'icutype':'icutype2'})
        
        return X_df
    
    def transform(self, union_array, *args):
        
        df = self.array_to_df(union_array)
        
        return df


    def fit(self, union_array, *args):
        return self
```

<div style="width:900px;background:#F9EECF;border:1px solid black;text-align:left;padding:20px;">

<b> Variance Inflation Factor</b>

</div>

As we can see from the correlation matrix that there are some features that are multicollinear with each other. The existence of multicolinear features may add noise to our model resulting in overfitting. Using the variance inflation factor(VIF), we will be able to estimate how much the variance of a regression coefficient of a particular feature is inflated due to multicollinearity.
<br><br>
The VIF ranges from 1 (1 being not correlated). We have set a threshold of 5, any features with VIF > 5 will be eliminated from the modeling.
<br><br>
$$VIF = \frac{1}{1 - R_{i}^{2}}$$
where $i = x_{1},x_{2}..x_{n}$

```python
# Variance Inflation Factor

class VIF(BaseEstimator, TransformerMixin):
    """Variance Inflation Factor"""
    
    def __init__(self,threshold=5):
        self.threshold = threshold
    
    
    def vif(self,df):
        
        vif = [variance_inflation_factor(df.iloc[:,:-4].values,i) \
               for i in range(df.iloc[:,:-4].shape[1])]
        
        vif_df = pd.DataFrame(df.iloc[:,:-4].columns,columns=['Features'])
        vif_df['VIF'] = vif    # VIF values in dataframe
        
        remove_col = list(vif_df[vif_df['VIF']>self.threshold]['Features'])   # Choose only features with VIF < 5
        selected_df = df.drop(remove_col,axis=1)
        
        return selected_df
    
    def transform(self, df, *args):
        
        selected_df = self.vif(df)
        
        return selected_df


    def fit(self, df, *args):
        return self
```

```python
feature_union = FeatureUnion([  
    ('numerical_scaler',Numerical_Extractor()),
    ('categorical_transform',Categorical_Extractor())  ])

pipeline = Pipeline(
    
    steps=[ ('drop_data', feature_union),
           ('clean_union',Clean_FeatureUnion()),
           ('VIF',VIF())
          
          ])

X_transformed = pipeline.fit_transform(X)

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.33, random_state=2,stratify=y)
```

```python
# Fit resampled training set on Logistic Regression CV

logregcv = LogisticRegressionCV(n_jobs=-1, random_state=6,max_iter=200)
logregcv.fit(X_train_res, y_train_res)

y_pred = logregcv.predict(X_test)  # predictions on test

logreg_precision = precision_score(y_test,y_pred)    # precision score
logreg_recall = recall_score(y_test,y_pred)          # recall score

logreg_score = pd.DataFrame([['logistic_regression',logreg_precision,logreg_recall]],
                            columns=score_compare.columns)     # append scores in table

score_compare = score_compare.append(logreg_score)
score_compare
```

```python
from sklearn.model_selection import learning_curve


def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - An object to be used as a cross-validation generator.
          - An iterable yielding train/test splits.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
title = 'Learning Curve Random Forest'
plot_learning_curve(rfc_gs.best_estimator_, title, X, y, ylim=None, cv=None,\
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5))
```

There is a gap between the training and cross-validation scores and the scores are generally low. The gap implies that there is high variance in the data set. We can look into simplifying the model and reduce complexity.
