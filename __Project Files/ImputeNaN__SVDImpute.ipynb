<span style="color:purple;font-size:13pt"><b>Data Cleaning</span></b>
<br><br>
<b>Step 1:</b> Drop all columns with more than 25% NaN values
<br><b>Step 2:</b> Drop all rows with more than 50% NaN values
<br><b>Step 3:</b> Drop all other outcomes labels except 'in-hospital_death'
</div>

```python
class DropNaN(BaseEstimator, TransformerMixin):
    
    def __init__(self):
        pass
        
    
    def remove_columns_(self,df,drop_col=1000):
        """remove columns with more than 1000 NaN values"""
        
        null_col = df.isnull().sum()
        df = df.drop(null_col[null_col > drop_col].index,axis=1)
        
        return df
    
    def remove_rows_(self,df,drop_row=57):
        
        df['NaNs'] = df.isnull().sum(axis=1)
        df = df.drop(df[df['NaNs']>drop_row].index)
        
        return df
    
    def drop_deslabels_(self,df,deslabels=['saps-i','sofa','length_of_stay','survival']):
        
        df = df.drop(deslabels,axis=1)
        
        return df
    
    def transform(self, main_df, *args):
        
        main_df = self.remove_columns_(main_df)
        main_df = self.remove_rows_(main_df)
        main_df = self.drop_deslabels_(main_df)
        
        self.feature_names = main_df.columns
        
        return main_df

    def fit(self, main_df, *args):
        return self
```

```python
dropnan = DropNaN()
mortality = dropnan.transform(mortality)
```

<span style="color:purple;font-size:13pt"><b>SVDImpute</span></b>
<br><br>
This algorithm obtains a set of principle components of the x variable matrix (eigenvectors) that can be linearly combined to approximate the NaN values in the data set.

It should be noted that SVD can only be performed on complete matrices; therefore we originally substitute row
average for all missing values in X matrix.

We then utilize an expectation maximization method to arrive at the final estimate, as follows. 

Each missing value in X is estimated using the above algorithm, and then the procedure is repeated on the newly obtained matrix, until the total change in the matrix falls below the empirically determined threshold of 0.01.
</div>

```python
X = mortality.drop('in-hospital_death',axis =1)
X.set_index('recordid',inplace=True)
```

```python
# Standard Scale 

scaled_array = []

for col in X.columns:

    masked_X = np.ma.array(X[col], mask=np.isnan(X[col]))

    masked_mean = np.mean(masked_X)
    masked_std = np.std(masked_X)

    scaled_col = np.array((masked_X - masked_mean)/masked_std)

    scaled_array.append(scaled_col)

np.transpose(scaled_array)
Xn = pd.DataFrame(data=np.transpose(scaled_array),columns=X.columns,index=X.index)
```

```python
X_filled_svd = IterativeSVD().complete(Xn)
```

```python
X_filled_svd = pd.DataFrame(data=X_filled_svd,columns=Xn.columns,index=Xn.index)
```
