```python
# Import packages

import pandas as pd
import numpy as np


import seaborn as sns

from sklearn.preprocessing import RobustScaler, OneHotEncoder
from sklearn.pipeline import FeatureUnion
from sklearn.base import BaseEstimator, TransformerMixin

from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from statsmodels.stats.outliers_influence import variance_inflation_factor

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows',200)


sns.set_style('whitegrid')
sns.set(rc={"figure.figsize": (15, 8)})

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
# Simple cleaning on csv file

mortality = pd.read_csv('mortality_filled_median.csv')
mortality.drop(['Unnamed: 0'],axis=1,inplace=True)
mortality.set_index('recordid',inplace=True)
```

```python
X = mortality.drop('in-hospital_death',axis=1)   # identify predictors 
```

```python
y = mortality['in-hospital_death']
```

<p>

<span style="color:purple;font-size:13pt"><b>Feature Selection</span></b>
<br><br>
<b>Step 1: Preprocess</b>
<br>
Get dummy variables for categorical features
<br>
Scale numerical features with Robust Scaler 
<br>
(Scaled with median and the interquartile range so that outliers can be taken into account)
<br><br>
<b>Step 2: Variance Inflation Factor</b> 
<br>Drop columns that are multi-colinear with each other
<br><br>
<b>Step 3: Decision Trees on Feature Importance</b>
<br> Rank features based on importance
<br><br>
<b>Step 4: Reduce dimensionality using PCA</b>
</div>

```python
class Categorical_Extractor(BaseEstimator, TransformerMixin):
    """Get dummy variables for categorical features"""
    
    def __init__(self):
        pass

    def get_dummy_(self,df):
        
        icu_dummy = pd.get_dummies(df['icutype'],drop_first=True)
        df = pd.concat((df['gender'],icu_dummy),axis=1)
        
        return df.values.reshape(-1, 4)
    
    def transform(self, X, *args):
        
        X = self.get_dummy_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
class Numerical_Extractor(BaseEstimator, TransformerMixin):
    """Use RobustScaler on numerical features"""
    
    def __init__(self):
        pass

    def robust_scale_(self,df,threshold=10):
        
        num_col = []
        for col, num in df.nunique().iteritems():
            if num > threshold:
                num_col.append(col)
                
        df = df[num_col]
        
        rscaler = RobustScaler()
        array = rscaler.fit_transform(df)
        
        return array.reshape(-1,array.shape[1])
    
    def transform(self, X, *args):
        
        X = self.robust_scale_(X)
        
        return X


    def fit(self, X, *args):
        return self
```

```python
feature_union = FeatureUnion([  
    ('numerical_scaler',Numerical_Extractor()),
    ('categorical_transform',Categorical_Extractor())
])
```

```python
transformed_X = feature_union.transform(X)
```

```python
class Clean_FeatureUnion(BaseEstimator, TransformerMixin):
    """Convert array from feature union into dataframe"""
    
    def __init__(self):
        pass
    
    
    def array_to_df(self,array):
        
        new_columns = list(X.columns)
        new_columns.extend(['icutype3','icutype4'])
        
        X_df = pd.DataFrame(array,columns=new_columns,index=X.index)
        X_df = X_df.rename(columns={'icutype':'icutype2'})
        
        return X_df
    
    def transform(self, union_array, *args):
        
        df = self.array_to_df(union_array)
        
        return df


    def fit(self, union_array, *args):
        return self
```

```python
clean_FU = Clean_FeatureUnion()
transformed_X = clean_FU.transform(transformed_X)
```

```python
# Variance Inflation Factor

vif = [variance_inflation_factor(transformed_X.iloc[:,:-4].values,i) 
       for i in range(transformed_X.iloc[:,:-4].shape[1])]
```

```python
sns.set_style('whitegrid')
sns.set(rc={"figure.figsize": (15, 15)})

vif_df = pd.DataFrame(transformed_X.iloc[:,:-4].columns,columns=['Features'])
vif_df['VIF'] = vif

sns.barplot(x='VIF',y='Features',data=vif_df)
```

```python
remove_col = list(vif_df[vif_df['VIF']>5]['Features'])
```

```python
selected_X = transformed_X.drop(remove_col,axis=1)
selected_X.shape
```

```python
# Variance Inflation Factor

class VIF(BaseEstimator, TransformerMixin):
    """Variance Inflation Factor"""
    
    def __init__(self):
        pass
    
    
    def vif(self,df):
        
        vif = [variance_inflation_factor(df.iloc[:,:-4].values,i) \
               for i in range(df.iloc[:,:-4].shape[1])]
        
        vif_df = pd.DataFrame(df.iloc[:,:-4].columns,columns=['Features'])
        vif_df['VIF'] = vif    # VIF values in dataframe
        
        remove_col = list(vif_df[vif_df['VIF']>5]['Features'])   # Choose only features with VIF < 5
        selected_df = df.drop(remove_col,axis=1)
        
        return selected_df
    
    def transform(self, df, *args):
        
        selected_df = self.vif(df)
        
        return selected_df


    def fit(self, union_array, *args):
        return self
```

```python
# Decision Tree Feature Importance

# Gridsearch Params
dtc_params = {
    'max_depth':[None,1,2,3,4],
    'max_features':[None,'log2','sqrt',2,3,4,5],
    'min_samples_split':[2,3,4,5,10,15,20,25,30,40,50]
}


dtc_gs = GridSearchCV(DecisionTreeClassifier(), dtc_params, cv=5, verbose=1)
dtc_gs.fit(selected_X, y)
```

```python
dtc_best = dtc_gs.best_estimator_
print dtc_gs.best_params_
print dtc_gs.best_score_
```

```python
fi = pd.DataFrame({
        'feature':selected_X.columns,
        'importance':dtc_best.feature_importances_
    })

fi.sort_values('importance', ascending=False, inplace=True)
fi
```

```python
# RandomForestClassifier Feature Importance

rfc_params = {
    'n_estimators':[5,10,15,20],
    'max_depth':[None,1,2,3,4],
    'max_features':[None,'log2','sqrt',2,3,4,5],
    'min_samples_split':[2,3,4,5,10,15,20,25,30,40,50]
}


rfc_gs = GridSearchCV(RandomForestClassifier(), rfc_params, cv=5, verbose=1)
rfc_gs.fit(selected_X, y)
```

```python
rfc_best = rfc_gs.best_estimator_
print rfc_gs.best_params_
print rfc_gs.best_score_
```

```python
rffi = pd.DataFrame({
        'feature':selected_X.columns,
        'importance':rfc_best.feature_importances_
    })

rffi.sort_values('importance', ascending=False, inplace=True)
rffi
```
